```python
df_diag_m = df_cancer[df_cancer['diagnosis']=='M']
df_diag_b = df_cancer[df_cancer['diagnosis']=='B']
```

```python
df_diag_b['area_mean'].describe()
df_diag_m['area_mean'].describe()
```

```python
'The mean on malignant tumors is {0:.4f} while that of benign tumors is {1:.4f}'.format(df_diag_m['area_mean'].mean(),df_diag_b['area_mean'].mean())
```

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.hist(df_diag_b['area_mean'],alpha=0.5,label='benign')
ax.hist(df_diag_m['area_mean'],alpha=0.5,label='malignant')
ax.set_title('Distribution of malignant and benign tumors in area_mean')
ax.set_xlabel('Area')
ax.set_ylabel('Count')
ax.legend(loc='upper right')
plt.show()

# mean_area of malignant tumors is generally larger than compared to bening tumor
```
